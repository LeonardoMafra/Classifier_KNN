```python
#usando biblioteca para calcular matriz de confusao sklearn


mt.confusion_matrix(y_train, y_pred)
```

```python
#funcao para calcular acuracia

acuracia = (mt.accuracy_score(y_train, y_pred)*100)
```

```python
print(f'Acuracia: {acuracia:.2f}')
```

```python
#prescisao

precisao = mt.precision_score(y_train, y_pred, pos_label = 'Conceder')
print(precisao)
```

```python
#recall

recall = (mt.recall_score(y_train, y_pred, pos_label = 'Conceder'))
print(recall)
```
